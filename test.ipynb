```python
import os
import openai
from langchain import OpenAI
from langchain.chains import RetrievalQA
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import Chroma
from langchain.document_loaders import DirectoryLoader, TextLoader, PyPDFLoader

# Load OpenAI API key from environment variable
openai.api_key = os.getenv("OPENAI_API_KEY")

# Step 1: Import the documents from the database (text and PDFs)
documents_directory = '/Users/vedantjoshi/Desktop/cq_Bot/ContractQuery'   # Replace with the path to your documents directory

# Use DirectoryLoader to load all text files and PDFs in the directory
text_loader = DirectoryLoader(documents_directory, glob="*.txt", loader_cls=TextLoader)
pdf_loader = DirectoryLoader(documents_directory, glob="*.pdf", loader_cls=PyPDFLoader)

# Load documents
text_documents = text_loader.load()
pdf_documents = pdf_loader.load()

# Combine all documents
documents = text_documents + pdf_documents

# Check for empty documents
valid_documents = [doc for doc in documents if doc.page_content.strip()]

# Generate unique IDs for each document
ids = [f"doc_{i}" for i in range(len(valid_documents))]

# Extract the texts and metadata
texts = [doc.page_content for doc in valid_documents]
metadatas = [doc.metadata for doc in valid_documents]

# Step 2: Initialize the OpenAI embedding model
embeddings = OpenAIEmbeddings()

# Step 3: Initialize ChromaDB and add documents
vectorstore = Chroma(persist_directory="chromadb_data", embedding_function=embeddings)
vectorstore.add_texts(texts=texts, metadatas=metadatas, ids=ids)

# Persist the ChromaDB instance
vectorstore.persist()

# Step 4: Create the RetrievalQA chain using Langchain
retrieval_qa = RetrievalQA.from_chain_type(
    llm=OpenAI(temperature=0),  # OpenAI model instance
    chain_type="stuff",  # RAG retrieval strategy
    retriever=vectorstore.as_retriever()
)

# Step 5: Sample Queries for Testing
queries = [
    "What is the payment term mentioned in the contract?",
    "Who is the responsible party for breach of contract?",
    "What is the duration of the contract?",
    "What are the termination conditions mentioned?",
    "Who are the signatories of the contract?"
]

# Test the RAG system with the sample queries
for query in queries:
    result = retrieval_qa.run(query)
    print(f"Query: {query}")
    print(f"Answer: {result}")
    print("-" * 50)
```
